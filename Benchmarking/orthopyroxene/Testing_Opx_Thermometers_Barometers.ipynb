# This notebook compares the results from Thermobar to those calculated using the spreadsheet hosted on Keith Putirka's personal website
(http://www.fresnostate.edu/csm/ees/faculty-staff/putirka.html)
- Note, this may be different from the version on your computer, as pre-2021, the spreadsheet online calculated the fraction of P2O5 incorrectly. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Thermobar as pt
pd.options.display.max_columns = None
```

```python
out=pt.import_excel('Python_Liq_OpxBarometers_Test.xlsx', sheet_name="Sheet1")
my_input=out['my_input']
myOPXs1=out['Opxs']
myLiquids1=out['Liqs']
```

# Solving for pressure using temperature inputted by user in column "T_K"

## Putirka 2008 Equation 29a

```python
P_Put=my_input['P_FixedT_Put_eq29a']
P_py=pt.calculate_opx_liq_press(opx_comps=myOPXs1, liq_comps=myLiquids1, 
                    equationP="P_Put2008_eq29a", T=my_input['T_K'])
fig, ((ax1, ax2)) = plt.subplots(1, 2, figsize=(12, 5))
#plotting 1:1 line
ax1.plot([np.min(P_Put), np.max(P_Put)], [np.min(P_Put), np.max(P_Put)], '-r')
ax1.plot(P_py, P_Put, 'ok')
ax1.set_xlabel('P Python (kbar)')
ax1.set_ylabel('P Put (kbar)')
ax2.plot(P_py, 1000*(P_py-P_Put), 'ok')
ax2.set_ylabel('P Python - P Put (bars)')
ax2.set_xlabel('P Python (kbar)')
plt.subplots_adjust(left=0.125, bottom=None, right=0.9, top=None, wspace=0.5, hspace=None)
```

# Cosolving pairs of equations 

## Temperature from Putirka 2008 equation 28a and Pressure from Putirka 2008 equation 29b

-excel returns error when you try to co-solve, so what I did here was paste in the pressure that python gives you from the cosolving, use the spreadsheet to get temperature to show that they are consistent

```python
P_Put=my_input['P_Cosolve_Eq28a_Eq29b']*10
T_Put=my_input['T_Cosolve_Eq28a_Eq29b'] # as Putirka spreasheet outputs in GPa
Python_calc=pt.calculate_opx_liq_press_temp(opx_comps=myOPXs1, liq_comps=myLiquids1, equationP="P_Put2008_eq29b", equationT="T_Put2008_eq28a", iterations=100, eq_tests=True)
P_py=Python_calc['P_kbar_calc']
T_py=Python_calc['T_K_calc']
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
#plotting 1:1 line
ax1.plot([np.min(T_Put), np.max(T_Put)], [np.min(T_Put), np.max(T_Put)], '-r')
ax2.plot([np.min(P_Put), np.max(P_Put)], [np.min(P_Put), np.max(P_Put)], '-r')
ax1.plot(T_py, T_Put, 'ok')
ax1.set_xlabel('T Python (K)')
ax1.set_ylabel('T Put (K)')
ax2.plot(P_py, P_Put, 'or')
ax2.set_xlabel('P Python (kbars)')
ax2.set_ylabel('P Put (kbars)')
ax3.plot(T_py, T_py-T_Put, 'ok')
ax3.set_xlabel('T Python (K)')
ax3.set_ylabel('T Python - T Put')
ax4.plot(P_py, 1000*(P_py-P_Put), 'or')
ax4.set_xlabel('P Python (kbar)')
ax4.set_ylabel('P Python - P Put (bars)')
plt.subplots_adjust(left=0.125, bottom=None, right=0.9, top=None, wspace=0.5, hspace=None)
```
