# Melt matching function
- Has 2 outputs, 1st output is an average P and T for each cpx, e.g., if cpx1 matches to 10 liquids, it averages the P and T. 
- The second output, is all matches, e.g., before any averaging in case you want that, e.g., Cpx1-liquid4, Cpx1-Liquid 5..... etc. 
- You can choose between various putirka equations here for P and T. This uses neave and equation 33. For the names of other equations, replace equationT= and equationP= with the names shown in the cartoon in bold



- Eq crit specifies what you want to use. Here, All is selected, so does Kd, DiHd, EnFs, CaTs. 
- The sigma is the threshold to accept tests at. E.g., 1 sigma uses values recomended by Neave (e.g., Delta DiHd=0.06, Delta EnFs=0.05, Delta CaTs=0.03. 
- For Kd, can also set KdMatch s oa set value, e..g,if say KdMatch="Masotta", uses Masotta Kd. By default, uses Putirka (T depenent eq 35). Could also set KdMatch=0.27 (e.g, Scruggs and Putirka). 
- Can also specify KdErr, by default, its 0.03, but could up to it 0.06 to consider more matches. 
- Specify Fe3FeT here. Be careful not to have lots of conflicting informatoin, e.g., don't enter FeO and Fe2O3 in your input if you want to change this, and don't have a Fe3FeT column in your input either

Some Examples
If you set KdMatch to 0.3,  and KdErr to 0.03, it uses 0.3 +-0.03 by default.
Uses the eq criteria of neave if sigma=1 etc

# This cell does the melt maching, using equation P= Neave 2017, Equation T =Putirka, 2008, 

```python
# For Fe3FeT=0, which is Keith's method (ignoring Fe3+)
#
melt_match_out=pw.calculate_Cpx_Liq_PT_melt_matching(Liq_Comps=myLiquids1, Cpx_Comps=myCPXs1, 
                                                equationP="P_Neave2017", equationT="T_Put2008_eq33", KdMatch="Putirka", KdErr=0.03,
                                                     Eq_Crit="All",  Fe3FeT_Liq=0.0, sigma=1)

AvMatches_Gleeson_NoFe3=melt_match_out['Averaged_PTs']
AllMatches_Gleeson_NoFe3=melt_match_out['All_PTs']
```

# You can check distribution of equilibrium tests in the output...

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))

ax1.hist(AllMatches_Gleeson_NoFe3['Delta_EnFs']);
ax1.set_xlabel('Delta EnFs')

ax2.hist(AllMatches_Gleeson_NoFe3['Delta_DiHd']);
ax2.set_xlabel('Delta DiHd')

ax3.hist(AllMatches_Gleeson_NoFe3['Delta_CaTs']);
ax3.set_xlabel('Delta CaTs')

ax4.hist(AllMatches_Gleeson_NoFe3['Delta_Kd_Put2008']);
ax4.set_xlabel('Delta_Kd_Put2008')
```

```python
# You can inspect data at any time by doing variablename.head()
AvMatches_Gleeson_NoFe3.head()
```

```python
# Doing same thing using Fe3/FeT=0.15, and sigma=2

melt_match_out2=pw.calculate_Cpx_Liq_PT_melt_matching(Liq_Comps=myLiquids1, Cpx_Comps=myCPXs1, 
                                                equationP="P_Neave2017", equationT="T_Put2008_eq33", Eq_Crit="All", Fe3FeT_Liq=0.15, sigma=2)

AvMatches_Gleeson_15Fe3=melt_match_out2['Averaged_PTs']
AllMatches_Gleeson_15Fe3=melt_match_out2['All_PTs']
```

# Plotting
-s everything is pandas, you can just do variablename['Column heading'] to pull out the variable of interest. 

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(12, 10))
ax1.invert_yaxis()
ax2.invert_yaxis()
# Subplot1
ax1.errorbar(AvMatches_Gleeson_NoFe3['Mean_T_K_calc']-273.15,  AvMatches_Gleeson_NoFe3['Mean_P_kbar_calc'],
             xerr=AvMatches_Gleeson_NoFe3['st_dev_T_K_calc'], yerr=AvMatches_Gleeson_NoFe3['st_dev_P_kbar_calc'],
             fmt='d', ecolor='k', elinewidth=0.8, mfc='cyan', ms=8, mec='k')

ax1.set_xlabel('Temperature (C)')
ax1.set_ylabel('Pressure (kbar)')
ax1.set_title('No Fe$^{3+}$')
# Subplot2
ax2.errorbar(AvMatches_Gleeson_15Fe3['Mean_T_K_calc']-273.15,  AvMatches_Gleeson_15Fe3['Mean_P_kbar_calc'],
             xerr=AvMatches_Gleeson_15Fe3['st_dev_T_K_calc'], yerr=AvMatches_Gleeson_15Fe3['st_dev_P_kbar_calc'],
             fmt='d', ecolor='k', elinewidth=0.8, mfc='red', ms=8, mec='k')
ax2.set_xlabel('Temperature (C)')
ax2.set_ylabel('Pressure (kbar)')
ax2.set_title('15% Fe$^{3+}$')

# Subplot3

ax3.hist(AvMatches_Gleeson_15Fe3['Mean_P_kbar_calc'], facecolor='red',  edgecolor='k')
ax3.hist(AvMatches_Gleeson_NoFe3['Mean_P_kbar_calc'], facecolor='cyan',  alpha=0.3, edgecolor='c')
ax3.hist(AvMatches_Gleeson_NoFe3['Mean_P_kbar_calc'], facecolor='none',  alpha=1, edgecolor='c')
ax3.set_xlabel('Pressure (kbar)')
ax3.set_ylabel('# of measurements')

ax4.hist(AvMatches_Gleeson_15Fe3['Mean_T_K_calc']-273.15, facecolor='red',  edgecolor='k')
ax4.hist(AvMatches_Gleeson_NoFe3['Mean_T_K_calc']-273.15, facecolor='cyan',  alpha=0.3, edgecolor='c')
ax4.hist(AvMatches_Gleeson_NoFe3['Mean_T_K_calc']-273.15, facecolor='none',  alpha=1, edgecolor='c')
ax4.set_xlabel('Temperature (C)')
ax4.set_ylabel('# of measurements')
```
